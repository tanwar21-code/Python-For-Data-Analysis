```python
# String formatting or Tempalating
# there are many ways of doing that nowadays
# Basic method
tempelate = "{0:.2f} {1:s} are worth US${2:d}"
# {0:.2f} = specifies that first element will be a float value with upto 2 deciaml numbers
# {1:s} = specifies that 2nd element will be a string
# {2:d} = specifies that 3rd argument will be an integer
tempelate.format(11.3,'Pesos',3)
```

```python
# there is another way of doing it as well
# Formatted string which came in recent pyhton update
amount = 25
rate = 80
currency = 'USD'
f"{amount} {currency} will be equal to {amount*rate} INR."
```
